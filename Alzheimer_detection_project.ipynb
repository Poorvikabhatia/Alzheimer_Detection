# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Preprocessing

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df.drop(['PatientID', 'DoctorInCharge','EducationLevel','BMI','Gender','Ethnicity','Smoking','Forgetfulness','Depression','Hypertension','CardiovascularDisease','Confusion','PersonalityChanges','Disorientation','Diabetes','DifficultyCompletingTasks','HeadInjury','FamilyHistoryAlzheimers'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['Diagnosis'])
y = df['Diagnosis']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.sample(5)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=150,criterion='gini')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

# Calculating classification_report, confusion_matrix, accuracy_score

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```
